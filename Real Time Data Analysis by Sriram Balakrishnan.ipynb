```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set(style="whitegrid")

# List of numerical columns (excluding categorical ones)
numerical_columns = ["Area", "Perimeter", "Major Axis Length", "Minor Axis Length", 
                     "Soma Diameter", "Aspect Ratio", "Circularity", "Solidity"]

# 🎨 Box Plots grouped by Cell Type
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size based on number of features
    sns.boxplot(data=df, x="Cell Type", y=col, palette="deep")
    plt.xticks(rotation=45)
    plt.title(f"Box Plot of {col} by Cell Type")

plt.tight_layout()
plt.show()

# 🎨 Histograms grouped by Cell Type
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data=df, x=col, hue="Cell Type", bins=20, kde=True, palette="deep", element="step")
    plt.title(f"Histogram of {col} by Cell Type")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# List of numerical columns (excluding categorical ones)SS
numerical_columns = ["Area", "Perimeter", "Major Axis Length", "Minor Axis Length", 
                     "Soma Diameter", "Aspect Ratio", "Circularity", "Solidity"]

# 🎨 KDE Plots grouped by Cell Type
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.kdeplot(data=df, x=col, hue="Cell Type", fill=True, common_norm=False, palette="deep", alpha=0.5)
    plt.title(f"KDE Plot of {col} by Cell Type")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for seaborn
sns.set(style="whitegrid")

# List of numerical columns (excluding categorical ones)
numerical_columns = ["Area", "Perimeter", "Major Axis Length", "Minor Axis Length", 
                     "Soma Diameter", "Aspect Ratio", "Circularity", "Solidity"]

# 🎨 Violin Plots grouped by Cell Type
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size based on number of features
    sns.violinplot(data=df, x="Cell Type", y=col, palette="deep")
    plt.title(f"Violin Plot of {col} by Cell Type")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot of all numerical features, differentiated by Cell Type
sns.pairplot(df, hue="Cell Type", vars=["Area", "Perimeter", "Major Axis Length", "Minor Axis Length", 
                                        "Soma Diameter", "Aspect Ratio", "Circularity", "Solidity"], palette="deep")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df[["Area", "Perimeter", "Major Axis Length", "Minor Axis Length", 
                  "Soma Diameter", "Aspect Ratio", "Circularity", "Solidity"]].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
